##__DateRange__
Date range is a sequence of datetime values with a fixed frequency. You can create a date range using the pd.date_range() function.

Key parameters of pd.date_range():
- __Start__: The start date of the range.
- __End__: The end date of the range.
- __Periods__: The number of periods to generate.
- __Freq__: The frequency of the date range (e.g., 'D' for daily, 'M' for monthly, 'H' for hourly).
- __.date_range__: A Date Range in Pandas is a sequence of date-time indices. It's created using the pd.
- __.dt.day__: To extract day from date.
- __.dt.month__: To extract month from date
- __.dt.year__: To extract year from month
- __.dt.weekday__: To get weekday in (0,1,2,3,4,5,6) where 0 = Monday, 1 = Tuesday and so onn...
- __pd.Timedelta(days=1)__: Timedelta represents a duration, the difference between two dates or times. We can perform variuos airthmetic operation using timedelta.

```python
import pandas as pd

date_range = pd.date_range(start='01-01-2024', end='01-01-2025', freq='m')
df = pd.DataFrame(date_range)
df
```

###Extracting Day, Month, & Year from Date.

```python
data = {'Date': ['01-Jan-2023','01-Feb-2023','01-Mar-2023','01-Apr-2023','01-May-2023','01-Jun-2023','01-Jul-2023','01-Aug-2023','01-Sep-2023','01-Oct-2023','01-Nov-2023','01-Dec-2023']}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

# Extracting day, month, and year information
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Displaying the DataFrame with extracted information
df[['Date','Day','Month','Year']]
```

###Getting Weekdays.

```python
data = {'Date': pd.date_range(start='2024-01-01', end='2024-01-31')}
week = pd.DataFrame(data)

#Getting Weekdays using .dt.weekday.
week['WeekDays'] = week['Date'].dt.weekday
week['WeekEnds'] = week['Date'].dt.weekday // 5 == 1 #30/5 == 1 (True)

week[['Date','WeekDays','WeekEnds']]
```

###Shifting dates forward or backward.

```python
date = {'Date': pd.date_range(start='2024-01-01', end='2024-01-31')}
df = pd.DataFrame(date)

#Adding or Substracting date using .Timedelta(days=1) function.
df['Previous_date'] = df['Date'] - pd.Timedelta(days=1)
df['Upcoming_date'] = df['Date'] + pd.Timedelta(days=1)

df[['Date','Previous_date','Upcoming_date']]
```
